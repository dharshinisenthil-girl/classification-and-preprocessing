```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load dataset (header row = feature names)
df = pd.read_csv("2.radiomics.csv")

# Separate feature names (exclude label)
feature_cols = df.columns[:-1]
label_col = df.columns[-1]

# Create scaler
scaler = MinMaxScaler()

# Scale features and create a NEW DataFrame (this avoids dtype issues)
X_scaled = pd.DataFrame(
    scaler.fit_transform(df[feature_cols]),
    columns=feature_cols
)

# Combine scaled features with label
df_normalized = pd.concat([X_scaled, df[label_col]], axis=1)

# Save to CSV
df_normalized.to_csv("2.radiomics.csv",index=False)

print("Normalization successful. File saved.")
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(len(iris.target))

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=43
)

# SVM model
model = SVC(kernel='poly')
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Results
print('Confusion matrix:\n',confusion_matrix(y_test, y_pred))
#print('Classification report:\n',classification_report(y_test, y_pred))
```
